```python
class Solution:

    def twoSum(self, nums: List[int], target: int) -> List[int]:
        numToIndex = {} # 字典
        for i in range(len(nums)):
            # 判斷鍵是否存在 Map
            if target - nums[i] in numToIndex: # 目標整數 - 索引 1 的值
                return [numToIndex[target - nums[i]], i]
            # 把前面的資料添加進字典
            numToIndex[nums[i]] = i # 鍵 nums[i] = 值 i
        return []
```
