```python
class Solution:
    def isCircularSentence(self, sentence: str) -> bool:
        if sentence[0]!=sentence[-1]:
            return False
        split=-1
        while True:
            split=sentence.find(' ', split+1)
            if split==-1:
                break
            if sentence[split-1]!=sentence[split+1]:
                return False
        return True
```
